```python
column_names = ['files', 'language', 'blank', 'comment', 'code', 'name']
df = pd.DataFrame(columns = column_names)
for repo in repo_name:
    try:
        test = pd.read_csv('codeline\\'+repo+'.csv')
        test = test.iloc[:, :-1]
        test['name']=repo
        df = pd.concat([df , test])
    except Exception as e:
        print(repo)
```

```python
print( "median number of LOC = {}".format(df.groupby('name').code.sum().median()))
```

```python
print( "median number of LOC = {}".format(df.groupby('name').language.sum().median()))
```
